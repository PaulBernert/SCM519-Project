```python
### DO NOT EDIT THIS CELL!!!
### DO NOT IMPORT ANY ADDITIONAL PACKAGES

import numpy as np
import tensorflow as tf

import matplotlib.pyplot as plt
%matplotlib inline

from tensorflow.keras import layers 
from tensorflow.keras import activations 
from tensorflow.keras import initializers 
from tensorflow.keras import losses 
from tensorflow.keras import metrics 
from tensorflow.keras import optimizers 
from tensorflow.keras import regularizers 
```

```python
### DO NOT EDIT THIS CELL!!!

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# read the breast cancer dataset from sklearn
dataset = load_breast_cancer()

# samples X: mxn
X = dataset.data
print ("X: ", X.shape)

# targets y: mx1
y = dataset.target
print ("y: ", y.shape)

# samples X: mxn
X = dataset.data
print ("X: ", X.shape)

# targets y: mx1
y = dataset.target
print ("y: ", y.shape)

# features
features = dataset.feature_names
print ("features: ", features)

# split the data into train and test
X_train, X_test, y_train, y_test = train_test_split(dataset.data, dataset.target, test_size=0.1)

# scale the data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
##### THIS IS THE ONLY CELL YOU WILL EDIT PER THE ISNTRUCTIONS BELOW

m, n = X_train.shape
print ("number of samples, m: ", m)
print ("number of features, n: ",n)

# BUILD sequential model
model = tf.keras.models.Sequential([
    
    # tell keras the input shape is (n,); 
    ### (1.1) YOUR CODE HERE
    #input_shape=(n,),
    
    # NO HIDDEN LAYERS!
    
    # create Dense output layer with 1 node & sigmoid activation (leave everything else at default values)
    ### (1.2) YOUR CODE HERE
    tf.keras.layers.Dense(1, activation='sigmoid', input_shape=(n,)),
     
])

# DISPLAY model architecture
# you should end up with 31 parameters
model.summary()

# COMPILE model
model.compile(
    
    # use SGD optimizer with learning rate 0.001 (leave everything else at default values)
    ### (1.3) YOUR CODE HERE
    tf.keras.optimizers.SGD(learning_rate=0.001),
    
    # use BinaryCrossentropy loss (leave everything at default values)
    ### (1.4) YOUR CODE HERE
    tf.keras.losses.BinaryCrossentropy(),
    
    # track accuracy metric
    ### (1.5) YOUR CODE HERE
    tf.keras.metrics.Accuracy(),
) 

# TRAIN model 
history = model.fit(
    
    # use X_train
    ### (1.6) YOUR CODE HERE
    x=X_train,
    
    # use y_train
    ### (1.7) YOUR CODE HERE
    y=y_train,
    
    # use 2500 epochs 
    ### (1.8) YOUR CODE HERE
    epochs=2500,
    
    # use batch size of m (which is equivalent to batch or deterministic gradient descent)  
    ### (1.9) YOUR CODE HERE
    batch_size=m,
    
    # use validation split of 0.1
    ### (1.10) YOUR CODE HERE
    validation_split=0.1,
) 
```

```python
##### DO NOT EDIT THIS CELL!!!

# PLOT loss and accuracy for train and validation data

# model.fit() returns a History object that contains a dictionary with everything that happened during training
# we can use it to plot training/validation loss and training/validation accuracy
history_dict = history.history
print (history_dict.keys())

# plot loss
loss = history_dict['loss']
val_loss = history_dict['val_loss']
epochs = range(1, len(loss) + 1)
plt.plot(epochs, loss, 'b', label='training loss')
plt.plot(epochs, val_loss, 'r', label='validation loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

# plot accuracy
acc = history_dict['accuracy']
val_acc = history_dict['val_accuracy']
epochs = range(1, len(acc) + 1)
plt.plot(epochs, acc, 'b', label='training accuracy')
plt.plot(epochs, val_acc, 'r', label='validation accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
##### DO NOT EDIT THIS CELL!!!

# EVALUATE model
# you should end up with a test accuracy of around 0.94-0.96
test_loss, test_acc = model.evaluate(X_test, y_test, verbose=2)
print ("\ntest set loss: ", round(test_loss,4), "\ntest set accuracy: ", round(test_acc,4))
```
